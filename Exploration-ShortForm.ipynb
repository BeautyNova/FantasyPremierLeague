```python
#Changing season to numerical value
df['season_x'] = np.where((df.season_x == '2016-17'),17, df.season_x)
df['season_x'] = np.where((df.season_x == '2017-18'), 18, df.season_x)
df['season_x'] = np.where((df.season_x == '2018-19'), 19, df.season_x)
df['season_x'] = np.where((df.season_x == '2019-20'), 20, df.season_x)
df['season_x'] = np.where((df.season_x == '2020-21'), 21, df.season_x)
df['season_x'] = np.where((df.season_x == '2021-22'), 22, df.season_x)
df.head()
```

```python
#creating variable to record average points in last 5 games
df['ShortForm'] = 0
```

```python
df['season_x'].dtype
```

```python
counter = 0
for p in df['ShortForm']:
    if df['season_x'].iloc[counter] != 17:
        if df['GW'].iloc[counter] == 1:
            season = df['season_x'].iloc[counter]
            name = df['name'].iloc[counter]
            GW = df['GW'].iloc[counter]
            df['ShortForm'].iloc[counter] = (df[(df['name'] == name) & (df['season_x'] == (season - 1)) & (df['GW'] > 33)]['total_points'].sum()) / 5
        elif df['GW'].iloc[counter] == 2:
            season = df['season_x'].iloc[counter]
            name = df['name'].iloc[counter]
            GW = df['GW'].iloc[counter]
            df['ShortForm'].iloc[counter] = ((df[(df['name'] == name) & (df['season_x'] == (season - 1)) & (df['GW'] > 34)]['total_points'].sum()) +
                                             df[(df['name'] == name) & (df['season_x'] == season) & (df['GW'] == 1 )]['total_points'].sum()) / 5
        elif df['GW'].iloc[counter] == 3:
            season = df['season_x'].iloc[counter]
            name = df['name'].iloc[counter]
            GW = df['GW'].iloc[counter]
            df['ShortForm'].iloc[counter] = ((df[(df['name'] == name) & (df['season_x'] == (season - 1)) & (df['GW'] > 35)]['total_points'].sum()) +
                                             df[(df['name'] == name) & (df['season_x'] == season) & (df['GW'] < 3 )]['total_points'].sum()) / 5
        elif df['GW'].iloc[counter] == 4:
            season = df['season_x'].iloc[counter]
            name = df['name'].iloc[counter]
            GW = df['GW'].iloc[counter]
            df['ShortForm'].iloc[counter] = ((df[(df['name'] == name) & (df['season_x'] == (season - 1)) & (df['GW'] > 36)]['total_points'].sum()) +
                                             df[(df['name'] == name) & (df['season_x'] == season) & (df['GW'] < 4 )]['total_points'].sum()) / 5
        elif df['GW'].iloc[counter] == 5:
            season = df['season_x'].iloc[counter]
            name = df['name'].iloc[counter]
            GW = df['GW'].iloc[counter]
            df['ShortForm'].iloc[counter] = ((df[(df['name'] == name) & (df['season_x'] == (season - 1)) & (df['GW'] == 38)]['total_points'].sum()) +
                                             df[(df['name'] == name) & (df['season_x'] == season) & (df['GW'] < 5 )]['total_points'].sum()) / 5
        else:
            season = df['season_x'].iloc[counter]
            name = df['name'].iloc[counter]
            GW = df['GW'].iloc[counter]
            df['ShortForm'].iloc[counter] = (df[(df['name'] == name) & (df['season_x'] == season) & (df['GW'] > (GW - 6)) & (df['GW'] < (GW))]['total_points'].sum()) / 5
    counter += 1
```
